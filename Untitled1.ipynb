```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import statsmodels.api as sm
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/numpy/numpy-tutorials/main/content/transistor_data.csv')
```

```python
data
```

```python
x = data['Date of Introduction'].tolist()
```

```python
y = [math.log(c) for c in data['MOS transistor count']]
```

```python
plt.scatter(x,y)
```

```python
x1 = sm.add_constant(x)
```

```python
r = sm.OLS(y, x1).fit()
```

```python
r.summary()
```

```python
"{:,}".format(math.exp(r.predict(exog=[1,2020])))
```
